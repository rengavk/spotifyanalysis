Spotify artist playist popularity analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("C:/Users/DELL/Downloads/8000_popular_tracks.csv")
df.head()
```

```python
#popularity distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
top_artists = df['Artists'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_artists.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Artists with Most Tracks')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df, vars=['Popularity'], hue='Artists')
plt.suptitle('Relationship between Popularity and Other Features')
plt.show()
```

```python
avg_popularity = df['Popularity'].mean()
print(avg_popularity)
```

```python
most_popular_track = df[df['Popularity'] == df['Popularity'].max()]
print(most_popular_track)
```

```python
avg_popularity_by_artist = df.groupby('Artists')['Popularity'].mean().sort_values(ascending=False).head(10)
highest_avg_popularity_artist = avg_popularity_by_artist.idxmax()
print(highest_avg_popularity_artist)
```
